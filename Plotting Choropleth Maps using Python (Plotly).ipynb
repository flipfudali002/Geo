````python
import json
import numpy as np
import pandas as pd
import plotly.express as px
````

````python
import plotly.io as pio
pio.renderers.default = 'chrome'
````

````python
india_states = json.load(open('/Users/apple/Downloads/states_india.geojson','r'))
````

````python
state_id_map = {}
for feature in india_states['features']:
    feature['id'] = feature['properties']['state_code']
    state_id_map[feature['properties']['st_nm']] = feature['id']
````

````python
df['Density'] = df['Density[a]'].apply(lambda x: int(x.split("/")[0].replace(",","")))
````

````python
df['id'] = df['State or union territory'].apply(lambda x: state_id_map[x])
````

````python
df.head()
````

````python
fig = px.choropleth(df, locations='id',geojson=india_states,color='Density')
fig.show()
````

````python
fig = px.choropleth(df, locations='id',geojson=india_states,color='Density',scope='asia')
fig.show()
````

````python
df['Density']
````

````python
df['DensityScale'] = np.log10(df['Density'])
#con este metodo de numpy pretendemos 
#acentuar las diferencias en las tonalidades del mapa 
df['DensityScale']
````

````python
fig = px.choropleth(df, locations='id',geojson=india_states,color='DensityScale',scope='asia')
fig.update_geos(fitbounds="locations",visible=False)
fig.show()
````

````python
fig = px.choropleth(df, locations='id',geojson=india_states,color='DensityScale')
fig.update_geos(fitbounds="locations",visible=False)
#esta linea de arriba deja solo india 
fig.show()
````

````python
fig = px.choropleth(df, 
                    locations='id',
                    geojson=india_states,
                    color='DensityScale',
                    hover_name = 'State or union territory',
                    hover_data = ['Density'])
fig.update_geos(fitbounds="locations",visible=False)
fig.show()
````

````python
fig = px.choropleth_mapbox(df,
                    locations = 'id',
                    geojson = india_states,
                    color = 'DensityScale',
                    hover_name = 'State or union territory',
                    hover_data = ['Density'],
                    mapbox_style = "carto-positron",
                    center = {'lat':24,'lon':78},
                    zoom = 3)
fig.show()
````

````python
fig = px.choropleth_mapbox(df,
                    locations = 'id',
                    geojson = india_states,
                    color = 'DensityScale',
                    hover_name = 'State or union territory',
                    hover_data = ['Density'],
                    mapbox_style = "carto-positron",
                    center = {'lat':24,'lon':78},
                    zoom = 3, opacity = 0.5)
fig.show()
````
